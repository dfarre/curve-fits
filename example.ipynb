```python
import random

from tcp_fits import example_functions, profiles
from tcp_fits.example_functions import MaxPartitionDiffV2
```

```python
def random_ints_array(max_abs=100, size=10):
    return random.choices(range(-max_abs, max_abs), k=size)

def make_comparison(series_length):
    return profiles.OneArgComparison(
        [random_ints_array(size=size) for size in range(2, series_length + 2)],
        MaxPartitionDiffV2, 'zero', 'one', index_name='Array size (-2)')
```

```python
comp = make_comparison(2000)
comp.polyfit(2)
comp.show(figsize=(15,5))
```
